## Example Multivariate SVM hinge loss

```python
def L_i(x, y, W):
    """
    unvectorized version. Compute the multiclass svm loss for a single 
    example (x,y)
    - x is a comulmn vector representing an image (e.g. 3073 x 1 in CIFAR-10)
      with an appended bias dimension in the 3073-rd position (i.r. bias trick)
    - y is an integer giving index of correct class (e.g. between 0 and 9 in CIFAR-10)
    -W is the weight matrix (e.g. 10 x 3073 in CIFAR-10)
    """
    delta = 1.0 # margin
    scores = W.dot(x) # scores of size 10 x 1, the scores for each class
    correct_class_score = scores[y]
    D = W.shape[0] # number of classes, e.g. 10
    loss_i = 0.0
    for j in xrange(D): # iterate over all wrong classes
        if j == y:
            # skip for the true class to only loop over incorrect classes
            continue
        # accumulate loss for the i-th example
        loss_i += max(0, scores[j] - correct_class_score + delta)
    return loss_i
```

```python
def L_i_vectorized(x, y, W):
    """
    Faster, half-vectorized implementation. Half-vectorized refers to the
    fact that for a single example the implementation contains no for loops,
    but there is still one loop over all the examples (outside this func)
    """
    delta = 1.0
    scores = W.dot(x)
    # compute the margins for all classes in one vector operation
    margins = np.maximum(0, scores - scores[y] + delta)
    # on the y-th position scores[y] - scores[y] canceled and gave delta.
    # we want to ignore the y-th position and only consider margin on max 
    # wrong class
    margins[y] = 0
    loss_i = np.sum(margins)
    return loss_i
```

```python
def L(X, y, W):
    """
    fully-vectorized implementation: 
    - X holds all training examples as columns (e.g 3073 x 50,000 in CIFAR-10)
    - y is array of integers specifying correct class (e.g. 50,000-D array)
    - W are weights (e.g. 10 x 3073)
    """
    n = X.shape[1] # num training examples
    delta = 1.0
    scores = W.dot(X)
    correct_scores = scores[y, np.arange(n)] # use lists to index correct scores
    margins = np.maximum(0, scores - correct_scores + delta)
    margins[y, np.arange(n)] = 0
    L = np.sum(margins)
    return L
```

## Numerical Gradient

```python
def eval_numerical_gradient(f, x):
    """
    a naive implementation of numerical gradient of f at x
    - f should be a function that takes a single argument
    - x is the point (numpy array) to evaluate the gradient at
    """
    
    fx = f(x) # evaluate function value at original point
    grad = np.zeros_like(x)
    h = 0.00001
    
    # iterate over all indexes in x
    it = np.nditer(x, flags=['multi_index'], op_flags=['readwrite'])
    while not it.finished:
        
        # evaluate function at x + h
        ix = it.multi_index # first index (0, 0)
        old_value = x[ix] # value at first weight of x
        x[ix] = old_value + h # increment by h
        fxh = f(x) # evaluate f(x + h)
        x[ix] = old_value # restore to previous value (very important!)
        
        # compute the partial derivative
        grad[ix] = (fxh - fx) / h # numerical derivative
#         grad[ix] = (fxh + fx) - (fxh- fx) / (2*h) # the slope (centered difference)
        it.iternext() # step to next dimension
        
    return grad
```

```python
# forward pass

N = X_dev.shape[0] # 500
D = X_dev.shape[1] # 3073 dimensions (features)
C = 10 # num classes
delta = 1.0 # margin, one is a fine default
lambd = 1.0
X = X_dev
y = y_dev # 500 x 1
# input is N x D
# weights must be D x C (each feature needs a weight, each class needs a row of weights)
# random weight init (small numbers)
W = np.random.randn(D, C) * 0.0001 # 3073 * 10

# compute scores
scores = X.dot(W) # 500(N) x 10(C)

# correct scores
correct_scores = scores[np.arange(N), y].reshape(N,1) # index the correct y at each row 500(N) by 1

# hinge loss vectors
margins = np.maximum(0, scores - correct_scores + delta)
# set margins for correct scores to 0
margins[np.arange(N), y] = 0 # check these before you set them to zero, they should be equal to delta

# calculate data loss (1/N) * sum(margins)
data_loss = np.mean(margins)

# add regularization loss
reg_loss = np.sum(W ** 2) # sum of elementwise sqauared weights

# calc full loss
L = data_loss + 0.5 * lambd * reg_loss

# now we need to do a backward pass, calculate the gradients of the loss with respect to each W
# how does each weight change the output of the loss?
# each weight will have a gradient 3073 features 10 classes
dW = np.zeros_like(W)

# derivative of max gate, if > 0 gradient is 1 else 0
binary = margins
binary[margins > 0] = 1
col_sum = np.sum(binary, axis=1) # sum of each column to see how many scores contributed to the loss per training example
binary[np.arange(N), y] = - col_sum[np.arange(N)] # set each y index to negative number of scores contributed, the rest should be 1
dW = np.dot(X.T, binary) # must be 3073 by 10 so make dimensions match up!
dW /= N # divide by num training examples

# add gradient from regularization
dW += lambd * W
dW
```
